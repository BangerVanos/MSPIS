```python
from src.x_arithmetic_pipeline import ArithmeticPipeline
import pandas as pd
from random import randint
```

## Rank dependencies

```python
ranks = []
accs = []
effs = []
procs = []
seq_tacts = []
par_tacts = []
for i in range(1, 200):
    vec_1 = [randint(0, 63) for _ in range(i)]
    vec_2 = [randint(0, 63) for _ in range(i)]
    pipeline = ArithmeticPipeline(vec_1, vec_2, 6, 6)
    while pipeline.is_busy:
        pipeline.tact()
    ranks.append(i)
    acc = round((i * 6) / (pipeline.status['tacts_done']), 5)
    eff = round(acc / 6, 5)
    accs.append(acc)
    effs.append(eff)
    procs.append(6)
    seq_tacts.append(i * 6)
    par_tacts.append(pipeline.status['tacts_done'])
_dict = {'Ранг задачи (r)': ranks,
         'T1': seq_tacts,
         'Tn': par_tacts,
         'Количество этапов (n)': procs,
         'Коэффициент ускорения K(r)': accs,
         'Эффективность e(r)': effs}
rdf = pd.DataFrame(_dict)
```

## Stages dependencies

```python
ranks = []
accs = []
effs = []
procs = []
seq_tacts = []
par_tacts = []
for i in range(1, 200):
    vec_1 = [randint(0, 63) for _ in range(i)]
    vec_2 = [randint(0, 63) for _ in range(i)]
    pipeline = ArithmeticPipeline(vec_1, vec_2, 6, 6)
    while pipeline.is_busy:
        pipeline.tact()
    ranks.append(i)
    ranks.append(i)
    acc = round((i * 6) / (pipeline.status['tacts_done']), 5)
    eff = round(acc / 6, 5)
    accs.append(acc)
    accs.append(round((i * 6) / (i * 6) ,5))
    effs.append(eff)
    effs.append(round((i * 6) / (i * 6  * 1) ,5))
    procs.append(6)
    procs.append(1)
    seq_tacts.append(i * 6)
    seq_tacts.append(i * 6)
    par_tacts.append(pipeline.status['tacts_done'])
    par_tacts.append(i * 6)
_dict = {'Ранг задачи (r)': ranks,
         'T1': seq_tacts,
         'Tn': par_tacts,
         'Количество этапов (n)': procs,
         'Коэффициент ускорения K(n)': accs,
         'Эффективность e(n)': effs}
pdf = pd.DataFrame(_dict)
```
