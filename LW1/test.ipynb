```python
from src.arithmetic_pipeline import ArithmeticPipeline
import plotly.graph_objects as go
import random
from math import floor
```

```python
# Acceleration coefficient graphic (dependence on problem rank)
r = 20
stages_amount = 6
p1_x = []
p1_y = []
p2_x = []
p2_y = []
for i in range(2, 200):
    p1_x.append(i + 2)
    p2_x.append(i)
    k = (r * (i + 1)) / (r + (i + 1) - 1)
    r2 = (k * (i - 1)) / (i - k)
    p2_y.append(r2)
    p1_y.append(r)
fig = go.Figure()
fig.add_trace(
    go.Scatter(x=p1_x, y=p1_y, mode='markers', name='1',
               text=[f'n = {n}; r(n) = {y}' for n, y in enumerate(p1_y, 4)],
               textposition='bottom center')
)
fig.add_trace(
    go.Scatter(x=p2_x, y=p2_y, mode='markers', name='2',
               text=[f'n = {n}; r(n) = {y}' for n, y in enumerate(p2_y, 2)],
               textposition='bottom center')
)
fig.update_layout(title_text='Зависимость коэффициента ускорения от ранга задачи',
                  font=dict(size=36), width=1280, height=720)
fig.update_xaxes(title_text='Ранг задачи')
fig.update_yaxes(title_text='Коэффициент ускорения K(r)')
fig.show(autosize=True) 
```
